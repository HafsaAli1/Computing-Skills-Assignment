```python
import numpy as np
import matplotlib.pyplot as plt

# Using np.linspace to create an array of 200 equally spaced values between 0 and 2Ï€.
t = np.linspace(0, 2 * np.pi, 200)

# Use the parametric equations given in the problem to compute x(t) and y(t) values for each t.
x = 2 * np.sin(t) * np.exp(np.cos(t))
y = (-3/2) * np.cos(t) * np.exp(np.sin(2*t))

# Plot the curve
plt.plot(x, y, 'b', label='Parametric Curve')
plt.xlabel('x'), plt.ylabel('y')
plt.title('Plot of Parametric Function')
plt.legend(), plt.grid(True)
plt.show()

```
