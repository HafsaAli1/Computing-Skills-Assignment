```python
# My name: Hafsa Ali
# Student ID: 2313846

import random
import matplotlib.pyplot as plt

# 1) Change myID to your student ID
myID = "2313846"  

# 2) User input validation
while (varCount := int(input("Enter an integer (>= 50): "))) < 50:
    print("Number must be at least 50. Try again.")

# 3) randSeedNum
s_d = sum(int(d) for d in myID)  
random.seed(s_d)  

# 4) Random list creation
data = [random.randint(1, 200) for _ in range(varCount)]

# Compute statistics
data_sum = sum(data)
data_mean = data_sum / len(data)
data_min = min(data)
data_max = max(data)

# Print results
print(f"Sum: {data_sum}, Mean: {data_mean:.2f}, Min: {data_min}, Max: {data_max}")

# Plot histogram
plt.hist(data, bins=10, color='blue', alpha=0.7)
plt.xlabel("Value"), plt.ylabel("Frequency"), plt.title("Histogram of Random Data")
plt.grid(True)
plt.show()
```
