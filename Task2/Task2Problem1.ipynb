```python
import numpy as np
import matplotlib.pyplot as plt

# This is to define the function f(x)
def f(x):
    return np.exp(-x**2) + 3 * np.sin(x + np.pi / 4)

# This will generate the x values from -2π to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 450)

# Compute f(x) for each point
y = f(x)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b', label=r'$f(x) = e^{-x^2} + 3\sin(x + \frac{\pi}{4})$')  # 'b' for blue line
plt.xlabel('x'), plt.ylabel('f(x)')
plt.title(r'Plot of $f(x) = e^{-x^2} + 3\sin(x + \frac{\pi}{4})$')
plt.axhline(0, color='black', ls='--', lw=0.5)
plt.axvline(0, color='black', ls='--', lw=0.5)
plt.legend(), plt.grid(True)


# Show the plot
plt.show()


```
