```python
# My name: Hafsa Ali
# Student ID: 2313846


import random
import matplotlib.pyplot as plt

# 1) Change myID to your student ID
myID = "2313846"  # Replace with your actual student ID

# 2) User input validation
while True:
    try:
        varCount = int(input("Enter an integer (>= 50): "))
        if varCount >= 50:
            break
        print("Number must be 50 or higher. Try again.")
    except ValueError:
        print("Invalid input. Please enter an integer.")

# 3) randSeedNum
s_d = sum(int(d) for d in myID)  # Sum of digits in myID
random.seed(s_d)  # Corrected from 'ssd' to 's_d'

# 4) Random list creation
data = [random.randint(1, 200) for _ in range(varCount)]

# Compute statistics
data_sum = sum(data)
data_mean = data_sum / len(data)
data_min = min(data)
data_max = max(data)

# Print results
print(f"Sum: {data_sum}, Mean: {data_mean:.2f}, Min: {data_min}, Max: {data_max}")

# Plot histogram
plt.hist(data, bins=10, color='blue', alpha=0.6)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")
plt.grid(True)
plt.show()
```
