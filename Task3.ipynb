```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate 1,000 random numbers from a normal distribution
random_data = np.random.normal(loc=0, scale=1, size=1000)

# Plot the histogram
plt.hist(random_data, bins=30, density=True, color='blue', alpha=0.6)

# Calculate descriptive statistics
mean = np.mean(random_data)
median = np.median(random_data)
std_dev = np.std(random_data)
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")

# Fit a normal distribution curve and overlay on the histogram
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

# Add labels and title
plt.title('Histogram with Normal Distribution Fit')
plt.xlabel('Data Values')
plt.ylabel('Density')

# Show the plot
plt.show()






```
